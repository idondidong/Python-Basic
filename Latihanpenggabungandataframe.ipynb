```python
kiri = {
    'nama' : ['ana','ana','andik','riko'],
    'nama2' : ['ana','andik','ana','andik'],
    'A' : ['a0','a1','a2','a3'],
    'B' : ['b0','b1','b2','b3']
}
kanan = {
    'nama' : ['ana','andik','andik','riko'],
    'nama2' : ['andik','andik','ana','riko'],
    'C' : ['c0','c1','c2','c3'],
    'D' : ['d0','d1','d2','d3']
}

dfkiri = pd.DataFrame(kiri)
dfkanan = pd.DataFrame(kanan)
```

```python
dfkiri
```

```python
dfkanan
```

```python
#outer
pd.merge(dfkiri, dfkanan, how='outer', on=['nama','nama2'])
```

```python
#inner
pd.merge(dfkiri, dfkanan, how='inner', on=['nama','nama2'])
```

```python
#left
pd.merge(dfkiri, dfkanan, how='left', on=['nama','nama2'])
```

```python
#righ
pd.merge(dfkiri, dfkanan, how='right', on=['nama','nama2'])
```
