```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.PairGrid(iris)
```

```python
gr = sns.PairGrid(iris)
gr.map(plt.scatter)
```

```python
gr = sns.PairGrid(iris)
gr.map_diag(plt.hist)
gr.map_upper(plt.scatter)
gr.map_lower(plt.k)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, palette='rainbow', hue='species')
```

```python
tips = sns.load_dataset('tips')
gr2 = sns.FacetGrid(tips, col='time', row='smoker')
gr2.map(plt.hist, 'total_bill')
```

```python
gr2 = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
gr2.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
gr = sns.JointGrid(x='total_bill', y='tip', data=tips)
```

```python
gr = sns.JointGrid(x='total_bill', y='tip', data=tips)
gr = gr.plot(sns.regplot, sns.distplot)
```
