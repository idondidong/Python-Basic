```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,10)
x
```

```python
y = x*2
y
```

```python
plt.plot(x,y, 'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Data Penjualan')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g--')
plt.subplot(1,2,2)
plt.plot(y,x, 'b*-')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y, 'y')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Ini Judul')

axes2.plot(x,y, 'y')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Ini Judul')
```

```python
fig = plt.figure(figsize=(8,4), dpi=150)
```

```python
fig, axes = plt.subplots(figsize=(8,4), dpi=150)
axes.plot(x,y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('JUDULKU')
```

```python
fig.savefig("ini plot saya.png", dpi= 150)
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,2,2])

axes.plot(x, x**2, label= "Data x pangkat 2")
axes.plot(x, x**3, label= "Data x pangkat 3")

axes.legend(loc=0)
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,2,2])

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=0.75)
axes.plot(x, x+4, color="red", linewidth=1)
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,2,2])

axes.plot(x, x+1, color="green", lw=2, linestyle='-')
axes.plot(x, x+2, color="green", lw=4, linestyle='-.')
axes.plot(x, x+3, color="green", lw=6, linestyle=':')
axes.plot(x, x+4, color="green", lw=8, linestyle='--')
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,2,2])

line, = axes.plot(x, x+1, color="black", lw=1.50)
line.set_dashes([5,5,3,5])
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,2,2])

axes.plot(x, x+1, color="green", lw=5, ls=':', marker='s', ms=10, markerfacecolor='y')
```
